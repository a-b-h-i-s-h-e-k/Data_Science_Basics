## Scatter Plot
Q. What is Scatter plot?
--> A scatter plot is a set of points that represents the value obtained for two different variablesplotted on a horizontal and vertical axes

#### When to use Scatter Plots?
--> Scatter plots are used to convey the relationship between two numerical variables

--> Scatter plots are sometimes called correlation plots because they show how two variables are correlated.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# index_col = 0 --> so that our default index value should be taken
# converting all the special characters into the nan values, we specify our all special characters that present in data
# na_values = ['??', '????']
# Now getting exact values i use method dropna, i am droping nan values, axis=0 -> so it will take rowwise,
# inplace = True -> so whatever the cells which are getting dropped should be implemented in the dataset

df = pd.read_csv('Toyota.csv', index_col = 0, na_values = ['??', '????'])
print(df.shape)
df.dropna(axis = 0, inplace = True)
print(df.shape)
```

```python
plt.scatter(df['Age'], df['Price'], c='blue')
plt.title("Scatter Plot of Price VS Age of the Cars")
plt.xlabel("Age")
plt.ylabel("Price")
plt.show()
```

## Histograms

<ul> Q. What is Histogram?
<li> It is graphical representation of data using bars of different heights.</li>
<li> Histogram groups numbers into ranges(number into ranges) and the height of each bar depicts the frequency
    (How many numbers count values) of each range or bin.</li>

</ul>

<ul> Q. When to use histograms?
<li> To represent the frequency distribution of numerical variables. </li>
</ul>

```python
plt.hist(df['KM'])

# we got plot but the ranges are not clear & bins are not fixed, we cannot specify the range paticularly because it si not fixed
# so there are additinal parameters which we can use.
```

```python
df.head()
```

```python
# defining edgecolor rangeses are clear, bins=5 means only 5 bars


plt.hist(df['KM'], color='green', edgecolor='white', bins=5)
plt.title("Histogram of Kilometer")
plt.xlabel('Kilometer')
plt.ylabel('Frequency')
plt.show()
```

## Bar Plot

<ul> what is bar plot>
    <li> A bar plot is a plot that represents categorical data with rectangular bars with lengths proportional to
    the counts that they represent</li>
    
</ul>

<ul> when to use bar plot?
    <li> To represent the frequency distribution of categorical variables.</li>
    <li> A bar diagram makes it easy to compare sets of data between different groups.</li>
    <li> When we have to check or we have to compare between different sets of groups, means if there are two or three
    categorical data or four or five categorical data in your particular variable or you can say column
    so for showing that which value is more which value is less which has been used more and which has use less we can use the bar plot.</li>
</ul>

```python
# we have to specify some variables

counts = [979,120,12]
fuelType = ('Petrol', 'Diesel', 'CNG')
# why we are creating an index => because it takes aarray
index = np.arange(len(fuelType))
# the above was the attributes means the parameters which we have prepared already, which we have to put inside the bar plot

plt.bar(index, counts, color=['red', 'blue', 'lightgreen'])
plt.title('Bar plot of FuelType')
plt.xlabel('FuelTypes')
plt.ylabel('Frequency')
plt.xticks(index, fuelType, rotation=90)
plt.show()
```

### Scatter Plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd. read_csv('Toyota.csv', index_col=0, na_values=['??', '????'])
print(df.shape)
df.dropna(axis=0,inplace=True)
print(df.shape)
```

```python
sns.set()
sns.regplot(x="Age",y="Price",data=df)
```

--> In graph we are seeing a line we call it regression fit line, that shows the incline of the flock, if we don't want this line we can use fit_reg= false so it don't give you regressor line.

--> If we change it to some other symbol instead of point in scatter plot we use marker to change in anything

--> Hue it is used to include another variable to show the fuel types of the categories.

```python
sns.set()
sns.regplot(x="Age", y="Price", data=df, fit_reg=False, marker = '+')
```

```python
sns.set()

sns.regplot(x="Age", y="Price", data=df, fit_reg=False, marker = '*')
# colour paletter = set1,2 for different colors
sns.lmplot(x="Age", y="Price", data=df, fit_reg=False, hue = 'FuelType', legend = True, palette = 'Set2')
```

## Histogram

```python
sns.histplot(df['Age'])
```

#### Why Use KDE?:
<ul>
<li> Smooth Representation: The histogram shows counts of data within bins, which can sometimes look blocky or inconsistent, especially with small datasets. The KDE adds a smooth curve that gives a better idea of the underlying distribution.</li>
    
<li> Probability Density: KDE provides a continuous estimate of the distribution, allowing you to understand where the data is concentrated and how it spreads, even if the histogram bins are coarse.</li>

<li> Non-Parametric: Unlike other distribution fits (like normal distribution), KDE does not assume any specific underlying statistical distribution. It just fits the data as smoothly as possible.</li>
</ul>

```python
# Kernel density estimate
sns.histplot(df['Age'], kde = True)
```

```python
sns.histplot(df['Age'], kde = True, bins = 6)
```

# Box and Whiskers Plot

Box and Whiskers plot - numerical variable

```python
sns.boxplot(y = df['Price'])
```

--> first whisker horizontal line showing lowest value and upper whisker horizontal line showing highest value and when we go upward in first wisker line we have another line in ox plot like the start line.

--> This box represnts the quartile for 25% middle one represents the 50% and third one represents the 75%. so they are also k/a first quartile, second quartile and third quartile.

--> first line in box plot represent as 25%(first quartile), second line in box plot represent as 50%(second quartile/median), third line in box plot represent as 75%(third quartile). 

--> As we can see some points upward the whisker these are some quartile, these are some outliers.

--> outliers means the values which are extreme highest

--> These outliers are at bottom also then means they are extreme lowest values

# Box and Whiskers plot for numerical vs categorical variable

```python
sns.boxplot(x = df["FuelType"], y = df["Price"])
```

# Grouped box and whiskers plot of Price VS FuelType and Automatic

```python
sns.boxplot(x = df["FuelType"], y = df["Price"], hue = df["Automatic"])
```

Let's plot box-whiskers plot and histogram on the same window

Split the plotting window into 2 parts

```python
f,(ax_box, ax_hist) = plt.subplots(2, gridspec_kw = {"height_ratios":(.50, .50)})
sns.boxplot(df["Price"], ax = ax_box)
sns.histplot(df["Price"], ax = ax_hist)
```
