# Detecting outliers using percentile

Quantile will give you the percentile value and if we want the data samples which are above 95% quantile<br>
9.68 is 95% percent quantile anything about this is something we can consider as an outliers<br>
what do you want to set your threshold to it really depend on sitatution, so if there is no liked fixed guideline<br>


```python
max_thresold = df['height'].quantile(0.95)
max_thresold
```

```python
# For detecting the outlier for high value
df[df['height']>max_thresold]
```

#

```python
min_thresold = df['height'].quantile(0.05)
min_thresold
```

```python
# outlier for  minimum value
df[df['height']<min_thresold]
```

using quantile is very useful beacuse what we are doing is we are removing the samples on the far ends onthe left far as well as right far, so for that reason quantile is one of the techniques and that we can use 

# Remove outliers

```python
df[(df['height']<max_thresold) & (df['height']>min_thresold)]
```
